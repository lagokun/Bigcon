```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
train=pd.read_csv("01_음식물쓰레기_FOOD_WASTE_210811_update.CSV", encoding = 'cp949')

train['em_kg']=train['em_g']/1000
train['em_t']=train['em_kg']/1000
```

```python
p1= pd.pivot_table(data=train, index='emd_nm', values='em_t',aggfunc='sum')
print(p1.sort_values(by='em_t',ascending=False).head(10))
```

```python
import warnings
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
warnings.filterwarnings('ignore')
```

```python
import matplotlib 

matplotlib.font_manager._rebuild()

plt.rc("font", family="Malgun Gothic")
```

```python
p1.sort_values(by='em_t',ascending=False).head(10)
top_list1=p1.sort_values(by='em_t',ascending=False).head(10).index.tolist()
sns.barplot(data=train, x='emd_nm', y='em_t' ,ci=False,order=top_list1,estimator=sum)
plt.title('TOP10 쓰레기 배출량 행정동') 
plt.xlabel('행정동')
plt.ylabel('쓰레기 배출량 / 톤기준')
plt.rc('font', family='NanumBarunGothic') 
plt.show()
```

```python
df1=pd.read_csv("01_음식물쓰레기_FOOD_WASTE_210811_update.CSV",encoding="cp949",low_memory=False)
df1['new date'] = pd.to_datetime(df1['base_date'])
df1['day']=df1['new date'].dt.day_name()
df1['month']=df1['new date'].dt.month
df1['year']=df1['new date'].dt.year
df1['em_kg']=df1['em_g']/1000
df1['em_t']=df1['em_kg']/1000
```

```python
import matplotlib 

matplotlib.font_manager._rebuild()

plt.rc("font", family="Malgun Gothic")
```

```python
sns.lineplot(data=df1, x='new date',y='em_t',estimator=sum,ci=None)
plt.title('제주도 쓰레기 배출량 추세') 
plt.xlabel('시기')
plt.ylabel('쓰레기 배출량 / 톤기준')
# plt.rc('font', family='NanumBarunGothic') 
plt.show()
```

```python
sns.lineplot(data=df1, x='day', y='em_t',estimator=sum,ci=None)
plt.title('요일별 쓰레기 배출량 추세') 
plt.xlabel('요일')
plt.ylabel('쓰레기 배출량 /톤기준')
# plt.rc('font', family='NanumBarunGothic') 
print(plt.show())
```

```python
sns.pointplot(data=df1, x='month',y='em_t',estimator=sum, ci=None, hue='year')
plt.title('달별 쓰레기 배출량 추세') 
plt.xlabel('달')
plt.ylabel('쓰레기 배출량 / 톤기준')
# plt.rc('font', family='NanumBarunGothic') 
print(plt.show())
```
